```
import pandas as pd
df = pd.DataFrame({'col1':[1,2,3,4],'col2':[444,555,666,444],'col3':['abc','def','ghi','xyz']})
df.head()
```

```
df['col2'].value_counts() # cuenta de valores únicos
```

### Funciones Apply

Funciones pre calculadas y propias aplicadas en forma broadcasting

```
def potencia2(x):
    return x**2
```

```
potencia2(34)
```

```
df['col1'].apply(potencia2)
```

```
df['col1'].apply(lambda x: x**2) # funciones anonimas (lambda)
```

```
df
```

```
df['col3'].apply(len)
```

```
df['col1'].sum()
```

### Tablas Pivot 
(niveles de indices)

```
data = {'A':['foo','foo','foo','bar','bar','bar'],
     'B':['one','one','two','two','one','one'],
       'C':['x','y','x','y','x','y'],
       'D':[1,3,2,5,4,1]}

df = pd.DataFrame(data)
```

```
df
```

```
df.pivot_table(values='D',index=['A', 'B'],columns=['C'])
```
