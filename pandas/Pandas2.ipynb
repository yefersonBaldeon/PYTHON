# Valores Faltantes

```
import numpy as np
import pandas as pd
```

```
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
```

```
df
```

```
df.dropna() # borra por defecto las filas con valores faltantes, requiere inplace = True
```

```
df
```

```
df.dropna(axis=1) # borra columnas con valores faltantes
```

```
df.dropna(thresh=2) # solo borra si mas de 2 valores faltantes
```

```
df.fillna(value='LLENAR') # imputación directa con un valor
```

```
df
```

```
df['A'].fillna(value=df['A'].mean()) # se puede llenar en función media, desviación etc.
```
