# Groupby


Group by permite agrupar filas de datos y aplicar funciones de agregación

```
import pandas as pd
# Crear DF
data = {'Compania':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Persona':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Ventas':[200,120,340,124,243,350]}
```

```
df = pd.DataFrame(data)
df
```

```
# Agrupar por Compañia
df.groupby('Compania')
```

```
porComp = df.groupby("Compania") # se puede grabar como variable
```

```
porComp
```

```
#Y llamar métodos agregación:
porComp.mean()
```

```
df.groupby('Compania').mean()
#df.groupby('Compania')['Ventas'].mean() 
```

```
#Más ejemplos de agregación:
porComp.std()
```

```
porComp.min()
```

```
porComp.max()
```

```
porComp.count()
```
