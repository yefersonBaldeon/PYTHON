```python
#-------------------------------------- MERGE => funciona con clave
```

```python
izquierda = pd.DataFrame({'clave': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
derecha = pd.DataFrame({'clave': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})  
```

```python
izquierda
```

```python
derecha
```

```python
izquierda.merge(derecha) # METODO 1
```

```python
pd.merge(derecha,izquierda) #METODO 2
```

```python
izquierda = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

derecha = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
izquierda
```

```python
derecha
```

```python
izquierda.join(derecha)
```
