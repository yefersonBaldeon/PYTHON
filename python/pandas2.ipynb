```python
import pandas as pd
```

```python
data = {'Compania':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Persona':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Ventas':[200,120,340,124,243,350]}

df=pd.DataFrame(data)

df
```

```python
df.groupby("Compania").count()
```

```python
df.groupby("Compania").count()[["Ventas"]]
```

```python
df.groupby("Compania").sum()
```
